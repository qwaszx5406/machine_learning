```python
import numpy as np
import matplotlib.pyplot as pt
x=np.arange(0,360)
y=np.sin(x*np.pi/180.0)
z=np.cos(x*np.pi/180.0)
pt.xlim(0,360)
pt.ylim(-1.2,1.2)
pt.title('XDD')
pt.plot(x,y,color='blue')
pt.plot(x,z,color='red')
pt.show
```

```python
import numpy as np
import matplotlib.pyplot as pt
x=np.arange(0,360)
y=np.sin(x*np.pi/180.0)
z=np.cos(x*np.pi/180.0)
pt.xlim(0,360)
pt.ylim(-1.2,1.2)
pt.title('XDD')
pt.plot(x,y,'blue',label='sin')
pt.plot(x,z, 'red',label='cos')
pt.xlabel('x')
pt.ylabel('y')
pt.legend()
pt.show
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
import numpy as np
from matplotlib import cm


def f(x, y):
    return np.cos(x**2 + y**2) / (1 + x**2 + y**2)

xgrid = np.linspace(-3, 3, 50)
ygrid = xgrid
x, y = np.meshgrid(xgrid, ygrid)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x,
                y,
                f(x, y),
                rstride=2, cstride=2,
                cmap=cm.jet,
                alpha=0.7,
                linewidth=0.25)
ax.set_zlim(-0.5, 1.0)
plt.show()
```
